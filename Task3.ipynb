```python
import os

dataset_path = '/content/drive/MyDrive/object detecction.v1i.yolov8-obb (2)/'  
for folder in ['train', 'valid', 'test']:
    images_path = os.path.join(dataset_path, folder, 'images')
    labels_path = os.path.join(dataset_path, folder, 'labels')
    print(f"{folder} images exist: {os.path.exists(images_path)}")
    print(f"{folder} labels exist: {os.path.exists(labels_path)}")
```

```python
from ultralytics import YOLO

# Load YOLO model
model = YOLO('yolov8n.pt')  

# Train the model
model.train(
    data= '/content/drive/MyDrive/object detecction.v1i.yolov8-obb (2)/data.yaml',  
    epochs=50,  
    imgsz=640  
)
```

```python
from ultralytics import YOLO

# Loading the model
model = YOLO('/content/runs/detect/train2/weights/best.pt') 

# Evaluation
metrics = model.val()  
# Print evaluation metrics
print(metrics)  
```
